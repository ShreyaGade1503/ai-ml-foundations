# Data Visualization using Matplotlib

This notebook demonstrates basic data visualization techniques using Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Student": ["Amit", "Sneha", "Rahul", "Priya", "Neha"],
    "Maths": [78, 85, 70, 88, 90],
    "Science": [82, 88, 75, 90, 92]
}

df = pd.DataFrame(data)
df
```

```python
plt.bar(df["Student"], df["Maths"])
plt.title("Maths Scores of Students")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```

```python
plt.plot(df["Student"], df["Maths"], marker='o', label="Maths")
plt.plot(df["Student"], df["Science"], marker='o', label="Science")

plt.title("Maths vs Science Scores")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

## Observations

- Priya and Neha perform consistently well in both subjects.
- Rahul has comparatively lower scores.
- Visualization helps identify performance patterns quickly.

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = {
    "Name": ["Amit", "Sneha", "Rahul", "Priya", "Neha"],
    "Age": [21, 22, 20, 23, 21],
    "Score": [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Score"].plot(kind="bar", title="Scores of Students")
plt.show()
```

## Observations

- The dataset contains basic student information.
- Scores range between 78 and 92.
- Priya and Neha have higher scores compared to others.
